```python
import os
from PIL import Image
```

function

```python
def convert_jpeg_to_jpg(input_dir, output_dir):
    # Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Iterate through all files in the input directory
    for filename in os.listdir(input_dir):
        if filename.lower().endswith('.jpeg'):
            # Construct full file path
            input_path = os.path.join(input_dir, filename)
            
            # Define the output filename
            output_filename = f"{os.path.splitext(filename)[0]}.jpg"
            output_path = os.path.join(output_dir, output_filename)
            
            # Open the image and convert it
            with Image.open(input_path) as img:
                img = img.convert('RGB')
                img.save(output_path, 'JPEG')
            print(f"Converted {filename} to {output_filename}")
```

```python
input_directory = './'  # Directory containing .jpeg files
output_directory = './0jpg'  # Directory to save .jpg files
```

```python
convert_jpeg_to_jpg(input_directory, output_directory)
```
